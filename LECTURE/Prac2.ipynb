# Конечномерная оптимизация

```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -np.exp(-np.square(x-0.7))
print(f(2))
```

```python
result = optimize.minimize_scalar(f)
```

```python
result.x
```

```python
(np.abs(result.x-0.7)/result.x)*100
```

```python
f(result.x)
```

```python
x = np.linspace(-1,2,500)
y = f(x)
```

```python
plt.plot(x,y)
```

```python
def d(x):
    return np.square(1 - x[0]) + 100 * (np.square(x[1] - np.square(x[0])))
```

```python
result = optimize.minimize(d, [0,0])
```

```python
result.x
```

```python
res = optimize.minimize(d,[0,0])
```

```python
def make_data_for_ploting_counter(d,x_lim,y_lim,resolution):
    x = np.arange(x_lim[0],x_lim[1],resolution)
    y = np.arange(y_lim[0],y_lim[1],resolution)
    
    xgrid,ygrid = np.meshgrid(x,y)
    z = d([xgrid,ygrid])
    return xgrid,ygrid,z
```

```python
x,y,z= make_data_for_ploting_counter(d, (-1.5,2),(-1,3),0.01)
```

```python
plt.contour(x,y,z,levels=20)
plt.scatter(res.x[0],res.x[1],c='b')
#plt.scatter(1,1,c='r')
```

```python
plt.scatter(res.x[0],res.x[1],c='b')
plt.scatter(1,1,c='r')
```

# Негладкая оптимизация

```python
np.random.seed(0)
```

```python
x = np.linspace(-5,5,101)
x_ = np.linspace(-5,5,31)
```

```python
def p(x):
    return -np.exp(-np.square(x))
#y_ = p(x_) + 0.2 *np.random.normal(size=31)
y_ = p(x_) + 0.2 *np.random.uniform(0,1)
plt.plot(x_,y_,linewidth=2,c='r')
plt.plot(x,p(x),linewidth=2,c='b')
```

```python
from scipy import interpolate
```

```python
p_inter = interpolate.interp1d(x_,y_)
```

```python
rezz = optimize.minimize_scalar(p_inter)
rezz
```

```python
plt.plot(x,p_inter(x))
plt.scatter(rezz.x,rezz.fun, c='r')
```
